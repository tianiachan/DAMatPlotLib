## Dependencies and starter code

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import seaborn as sns

# Study data files
mouse_metadata = "data/Mouse_metadata.csv"
study_results = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata)
study_results = pd.read_csv(study_results)
#put in data frame
mouse_metadata = pd.DataFrame(mouse_metadata)
study_results = pd.DataFrame(study_results)

# Combine the data into a single dataset
merged = pd.merge(study_results,mouse_metadata, how = "left")
merged
```

## Summary statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
#group by the drug regimen
mergedDF_grouped = merged.groupby(["Drug Regimen"])
#get the different stats and convert to lists
mean = mergedDF_grouped.mean()
mean = mean["Tumor Volume (mm3)"].tolist()
median = mergedDF_grouped.median()
median = median["Tumor Volume (mm3)"].tolist()
variance = mergedDF_grouped.var()
variance = variance["Tumor Volume (mm3)"].tolist()
stdDev = mergedDF_grouped.std()
stdDev = stdDev["Tumor Volume (mm3)"].tolist()
stdError = mergedDF_grouped.sem()
stdError = stdError["Tumor Volume (mm3)"].tolist()
#index = mergedDF_grouped.index.values.tolist()
mergedDF_grouped.head()
#place all stats in a summary dataframe
summary_stat_DF = pd.DataFrame({
     "Mean Tumor Volume": mean,
    "Median Tumor Volume": median,
    "Tumor Volume Variance":variance,
    "Tumor Volume Std. Dev.": stdDev,
    "Tumor Volume Std. Err.":stdError
 })
#add drug regimen column and make into an index
drug_regimen = mergedDF_grouped["Drug Regimen"].unique().tolist()
drug_regimen = ["Capomulin","Ceftamin","Infubinol","Ketapril","Naftisol","Placebo","Propriva","Ramicane","Stelasyn","Zoniferol"]
summary_stat_DF["Drug Regimen"] = drug_regimen
summary_stat_DF = summary_stat_DF.set_index("Drug Regimen")
summary_stat_DF
```

## Bar plots

## Quartiles, outliers and boxplots

```python
# Calculate the final tumor volume of each mouse across four of the most promising treatment regimens. Calculate the IQR and quantitatively determine if there are any potential outliers. 
#sort data frame by mouse ID and drug regimen 
sort_by_ID = merged.sort_values(["Mouse ID", "Drug Regimen"])
#get rid of the extra index that showed up
sort_by_ID = sort_by_ID.reset_index(drop=True)
#group by mouse ID and then get the last value using tail to get final tumor volumne
groupID = sort_by_ID.groupby(["Mouse ID"]).tail(1)


#get only capomulin final tumor volume
groupIDC = groupID.loc[groupID["Drug Regimen"] == "Capomulin"]
#get only capomulin IQR and quartiles
quartilesC = groupIDC.quantile([.25,.5,.75])
quartilesC
lowerqC = quartilesC.loc[0.25,'Tumor Volume (mm3)']
print(f"The lower quartile is : {lowerqC}")
upperqC = quartilesC.loc[0.75,'Tumor Volume (mm3)']
print(f"The upper quartile is : {upperqC}")
iqrC = upperqC-lowerqC
print(f"The interquartile range is : {iqrC}")
lower_boundC = lowerqC - (1.5*iqrC)
print(f"Values below {lower_boundC} could be outliers.")
upper_boundC = upperqC + (1.5*iqrC)
print(f"Values above {upper_boundC} could be outliers.\n")

#get only Ramicane
groupIDR = groupID.loc[groupID["Drug Regimen"] == "Ramicane"]
#get only ramicane IQR and quartiles
quartilesR = groupIDR.quantile([.25,.5,.75])
lowerqR = quartilesR.loc[0.25,'Tumor Volume (mm3)']
print(f"The lower quartile is : {lowerqR}")
upperqR = quartilesR.loc[0.75,'Tumor Volume (mm3)']
print(f"The upper quartile is : {upperqR}")
iqrR = upperqR-lowerqR
print(f"The interquartile range is : {iqrR}")
lower_boundR = lowerqR - (1.5*iqrR)
print(f"Values below {lower_boundR} could be outliers.")
upper_boundR = upperqR + (1.5*iqrR)
print(f"Values above {upper_boundR} could be outliers.\n")

#get only Infubinol
groupIDI = groupID.loc[groupID["Drug Regimen"] == "Infubinol"]
#get only Infubinol IQR and quartiles
quartilesI = groupIDI.quantile([.25,.5,.75])
lowerqI = quartilesI.loc[0.25,'Tumor Volume (mm3)']
print(f"The lower quartile is : {lowerqI}")
upperqI = quartilesI.loc[0.75,'Tumor Volume (mm3)']
print(f"The upper quartile is : {upperqI}")
iqrI = upperqI-lowerqI
print(f"The interquartile range is : {iqrI}")
lower_boundI = lowerqI - (1.5*iqrI)
print(f"Values below {lower_boundI} could be outliers.")
upper_boundI = upperqI + (1.5*iqrI)
print(f"Values above {upper_boundR} could be outliers.\n")


#get only Ceftamin
groupIDCe = groupID.loc[groupID["Drug Regimen"] == "Ceftamin"]
#get only Ceftamin IQR and quartiles
quartilesCe = groupIDCe.quantile([.25,.5,.75])
lowerqCe = quartilesCe.loc[0.25,'Tumor Volume (mm3)']
print(f"The lower quartile is : {lowerqCe}")
upperqCe = quartilesCe.loc[0.75,'Tumor Volume (mm3)']
print(f"The upper quartile is : {upperqCe}")
iqrCe = upperqCe-lowerqCe
print(f"The interquartile range is : {iqrCe}")
lower_boundCe = lowerqCe - (1.5*iqrCe)
print(f"Values below {lower_boundCe} could be outliers.")
upper_boundCe = upperqCe + (1.5*iqrCe)
print(f"Values above {upper_boundCe} could be outliers.\n")

```

## Line and scatter plots

```python
# Generate a line plot of time point versus tumor volume for a mouse treated with Capomulin
mousel509 = merged.loc[merged["Mouse ID"]== "l509"]
xValues = mousel509["Timepoint"]
yValues = mousel509["Tumor Volume (mm3)"]
plt.plot(xValues,yValues)
plt.title("Capomulin treatment of mouse l509")
plt.xlabel("Timepoint(days)")
plt.ylabel("Tumor Volume (mm3)")
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen
#get only capomulin data
capo = merged.loc[merged["Drug Regimen"]=="Capomulin"]
#get mean of tumor volume for each mouse
avgTumorCapo = capo.groupby(["Mouse ID"])["Tumor Volume (mm3)"].mean()
#weight does not change so can get mean of each mouse to get same amount of data points
mouseWeight = capo.groupby(["Mouse ID"])["Weight (g)"].mean()
#and graph it!
plt.scatter(mouseWeight,avgTumorCapo)
#label stuff
plt.title("Mouse Weight vs Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
```

```python
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
linRegress = st.linregress(mouseWeight,avgTumorCapo)
plt.scatter(mouseWeight,avgTumorCapo)
linRegressformula = mouseWeight*linRegress[0] + linRegress[1]
plt.plot(mouseWeight, linRegressformula)
plt.title("Linear Regression of Mouse Weight vs Tumor Volume")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
print(f"The correlation between mouse weight and the average tumor volume is {linRegress[2]}.")
```
